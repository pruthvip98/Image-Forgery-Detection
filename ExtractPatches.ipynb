```python
from os import listdir
from os.path import isfile, join
from PIL import Image
#import matplotlib.pyplot as plt
import numpy as np
from skimage.util import view_as_windows
import random
```

```python
mypath = 'C:\\Users\\DELL\\Desktop\\casia-dataset\\CASIA2\\Tp'
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
mypath2 = 'C:\\Users\\DELL\\Desktop\\casia-dataset\\CASIA2\\CASIA v2.0 Groundtruth'
onlyfilesT = [f for f in listdir(mypath2) if isfile(join(mypath2, f))]
```

```python
onlyfiles = np.sort(onlyfiles)
onlyfilesT = np.sort(onlyfilesT)
from skimage.color import rgb2ycbcr
```

```python
def ExtractPatchesPic(Source,wsz,stride):
    img = np.array(Image.open(Source))
    patches = view_as_windows(img,[32,32,3],1)
    patches = patches.reshape((patches.shape[0]*patches.shape[1],32,32,3))
    return patches 
```

```python
def ExtractPatches():
    
    for k in range(0,250):
        patches=[]
        output=[]
        for i in range(0,20):
            testPatches = ExtractPatchesPic("C:\\Users\\DELL\\Desktop\\casia-dataset\\CASIA2\\Tp\\" + onlyfiles[k*20 + i],32,1)
            testPatchesGT = ExtractPatchesPic("C:\\Users\\DELL\\Desktop\\casia-dataset\\CASIA2\\CASIA v2.0 Groundtruth\\" +  onlyfilesT[k*20 + i],32,1)
            print(k*20+i)
            for j in range(testPatches.shape[0]):
                inpPT = rgb2ycbcr(testPatches[j])[:,:,1:3]
                patches.append(inpPT)
                if(.35<np.sum(np.dot(testPatchesGT[j][:,:,:], [0.299, 0.587, 0.114])/255)/(32*32) 
                   and .65>np.sum(np.dot(testPatchesGT[j][:,:,:], [0.299, 0.587, 0.114])/255)/(32*32)):
                    output.append(1)
                else:
                    output.append(0)
        print(np.array(output).nonzero())
        dest= "batch" + str(k) + ".npy"
        dest1 = "batchOutput" + str(k) + ".npy"
        np.save(dest,patches)
        np.save(dest1,output)
```

```python
for i in range(0,50):
    ubatch = np.load("E:\\TampU\\UnTampBatch"+str(i)+".npy")
    zeros = np.zeros((ubatch.shape[0],1))
    tbatch = np.load("E:\\TampU\\TampBatch"+str(i)+".npy")
    ones = np.ones((tbatch.shape[0],1))
    ubatch = np.concatenate((ubatch,tbatch),axis=0)
    output = np.concatenate((zeros,ones),axis=0)
    
    a = ubatch
    b = output

    combined = list(zip(a, b))
    random.shuffle(combined)

    ubatch,output = zip(*combined)
    np.save("E:\\FinalBatch\\Batch"+str(i)+".npy",np.array(ubatch))
    np.save("E:\\FinalBatch\\Output"+str(i)+".npy",np.array(output))
```
